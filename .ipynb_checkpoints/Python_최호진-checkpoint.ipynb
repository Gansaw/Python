```python
### 1. import하기 ###

import requests
from bs4 import BeautifulSoup
```

```python
# 현재 진행하고 있는 메이플스토리 이벤트 제목 추출하기
# 메이플 인벤에서 크롤링을 진행

url = "http://maple.inven.co.kr"
response = requests.get(url)
print(response.text)
```

```python
### 2. 웹 크롤링 ###

url = "http://maple.inven.co.kr"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# 현재 진행하고 있는 메이플스토리 이벤트 정보
events = soup.find_all(class_="menuGroup bg event")

# 이벤트 제목 추출
title_list = []
for event in events:    
    titles = event.find_all("p", class_="ev_txt")
    for title in titles:
#         print(title.text)
        title_list.append(title.text)

print(title_list)

# 이벤트 날짜 추출
day_list = []
for event in events:
    days = event.find_all("p", class_="ev_day")
    for day in days:
#         print(day.text)
        day_list.append(day.text)
        
print(day_list)

# 이벤트 링크 추출
link_list = []
for event in events:
    links = event.find_all("a")
    for link in links:
        href = link.get("href")
#         print(href)
        link_list.append(href)

print(link_list)
```

```python
### 3. 데이터 전처리 ###

# 이벤트 날짜를 시작날짜와 종료날짜로 분리하기
# 이벤트 날짜를 날짜 형태로 변환하기(yyyy-MM-dd)

from datetime import datetime

dates = day_list
start_date_list = []
end_date_list = []

for date in dates:
    start_date, end_date = date.split("~")        
    start_date = datetime.strptime(start_date, "%m/%d").strftime("2023-%m-%d")
    end_date = datetime.strptime(end_date, "%m/%d").strftime("2023-%m-%d")
    print("Start Date:", start_date)    
    start_date_list.append(start_date)
    print("End Date:", end_date)
    end_date_list.append(end_date)
    
print(start_date_list)
print(end_date_list)

# 데이터프레임 제작

import pandas as pd

data = {
    "제목": title_list,
    "시작날짜": start_date_list,
    "종료날짜": end_date_list,
    "링크주소": link_list
}

df = pd.DataFrame(data)

print(df)

# CSV파일로 저장
df.to_csv('maple_events.csv', index=False)
```

```python
# 그래프를 그리기 전 한글폰트 적용 (한글이 깨질 수 있기 때문)
# 한글 폰트 적용 (PC마다 폰트 위치가 다른 경우 한글이 깨질 수가 있음)

from matplotlib import font_manager

# 맑은 고딕 폰트 경로 설정
font_path = "C:/Windows/Fonts/malgun.ttf"

# 폰트 설정
font_prop = font_manager.FontProperties(fname=font_path)
plt.rcParams["font.family"] = font_prop.get_name()
```

```python
### 4. 그래프 그리기 ###

# 그래프 1: Bar Graph

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7,7))

bars = ax.bar(df['제목'], (pd.to_datetime(df['종료날짜']) - pd.to_datetime(df['시작날짜'])).dt.days)
plt.xlabel('이벤트 제목')
plt.ylabel('이벤트 기간 (일)')
plt.title('이벤트 기간')
plt.xticks(rotation=45)

# 막대 위에 숫자 나타내기
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 숫자와 그래프 간 간격 지정
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Graph 2 : Linear Graph
# referred to chatbot

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
for i, row in df.iterrows():
    start_date = pd.to_datetime(row['시작날짜'])
    end_date = pd.to_datetime(row['종료날짜'])
    duration = (end_date - start_date).days
    plt.plot([start_date, end_date], [duration, duration], label=row['제목'])

plt.xlabel('날짜')
plt.ylabel('이벤트 기간 (일)')
plt.title('이벤트 기간')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # 범례 위치 조정
plt.xticks(rotation=45)
plt.show()
```
